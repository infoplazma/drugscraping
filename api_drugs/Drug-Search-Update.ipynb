<h2 style="color:blue">Обновление usage dosage:</h2>

#### Выставить значения:

```python
DRUG_ID = 2502
USAGE_COLUMN_NAME = 'Спосіб застосування та дози'
NEW_TEXT = get_usage_dosage_by_drug_id(IP, DRUG_ID, 'Спосіб застосування та дози')['text']
print(f"{NEW_TEXT=}")
```

#### Очистка текста при необходимости:

```python
NEW_TEXT = NEW_TEXT.replace('_x000D_', "").replace(r'\t', "").replace(r'\r', "").replace(r'\n', "")
NEW_TEXT = re.sub(r'\s+', " ", NEW_TEXT)
print(f"{NEW_TEXT=}")
```

### Выполнить обновление:

```python
try:
    pprint(update_usage_dosage(IP, DRUG_ID, USAGE_COLUMN_NAME, NEW_TEXT))
    print("="*120)
    print("Completed")
except Exception as e:
    print(e)
```
